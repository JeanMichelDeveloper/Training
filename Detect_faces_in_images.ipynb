```python
#import necessary libraries
import cv2
from google.colab.patches import cv2_imshow
```

```python
#download image
!wget "https://img.freepik.com/free-photo/girl-smiling-making-auto-photo-with-her-friends-around_1139-593.jpg?w=900&t=st=1671177843~exp=1671178443~hmac=5ae3a701c75b3e103745b64dac70de722eebf323f278bfa45c1c7036dc19faf9"
image_path = "girl-smiling-making-auto-photo-with-her-friends-around_1139-593.jpg?w=900&t=st=1671177843~exp=1671178443~hmac=5ae3a701c75b3e103745b64dac70de722eebf323f278bfa45c1c7036dc19faf9"
```

```python
#show image
image = cv2.imread(image_path)
cv2_imshow(image)
```

```python
#Get model
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
#convert image to gray scale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
#detect faces
scale_factor = 1.1
minimum_neighbors = 4

faces = face_cascade.detectMultiScale(gray_image, scale_factor, minimum_neighbors)

faces
```

```python
#draw faces on image
color = (255, 255, 0)
line_thickness = 2

for (x, y, width, height) in faces:
  cv2.rectangle(image,
                (x, y),
                (x + width, y + height),
                color,
                line_thickness)

cv2_imshow(image)
```
