```python
#import necessary libraries
import cv2
from google.colab.patches import cv2_imshow
import numpy
```

```python
!wget "https://cdn.pixabay.com/photo/2015/04/28/19/15/stop-sign-744192__480.jpg"
```

```python
image = cv2.imread("stop-sign-744192__480.jpg")

cv2_imshow(image)
```

```python
#convert to HSV
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

cv2_imshow(image_hsv)
```

```python
#determine mask variable
color = numpy.uint8([[[64,25,212]]]) #BRG of the color we want to isolate
hsv = cv2.cvtColor(color,cv2.COLOR_BGR2HSV)
print(hsv)

minimum_hsv = (172, 100, 0)
maximum_hsv = (176, 255, 255)
```

```python
#create mask
mask = cv2.inRange(image_hsv, minimum_hsv, maximum_hsv)

cv2_imshow(mask)
```
