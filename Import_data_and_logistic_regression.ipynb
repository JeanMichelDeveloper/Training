Import and clean data

```python
#load data
adult_path = '/content/drive/MyDrive/AI_resources/adult.data'

import pandas

data = pandas.read_csv(adult_path, header=None, index_col=False, names=['age', 'workclass', 'number', 'education', 'education-num', 
                                                                        'marital-status', 'occupation', 'relationship', 'race', 'gender', 
                                                                        'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'])
data.head()
```

```python
#select columns to be used
data = data[['age', 'workclass', 'education', 'gender', 'hours-per-week', 'occupation', 'income']]
data.head()
```

Build X and Y data with one hot encoding

```python
dummies = pandas.get_dummies(data)
dummies.head()
```

```python
#create features variables
features = dummies.loc[:,'age': 'occupation_ Transport-moving']
features.head()
```

```python
features.info()
```

```python
X = features.values
print(X)
```

```python
y = dummies['income_ >50K']
print(y)
```

Logistic regression with one hot encoding

```python
#import model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

#split data for training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

#fit model
model.fit(X_train, y_train)
```

```python
#evaluate model
print(model.score(X_test, y_test))
```

Scale and encode data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#create transformer
transformer = ColumnTransformer([
    ("scaling", StandardScaler(),["age", "hours-per-week"]),
    ("onehot", OneHotEncoder(sparse=False),["workclass", "education", "gender", "occupation"])
])
```

Build, train and test model

```python
#create features
features = data.drop("income", axis=1)
features.head()
```

```python
#split and transform train and test data
X_train, X_test, y_train, y_test = train_test_split(features, data.income, random_state = 0)
transformer.fit(X_train)
X_train_transformed = transformer.transform(X_train)
print(X_train_transformed)
```

```python
#create and fit model
model = LogisticRegression()
model.fit(X_train_transformed, y_train)
```

```python
#test model
X_test_transformed = transformer.transform(X_test)
accuracy = model.score(X_test_transformed, y_test)
print(accuracy)
```
