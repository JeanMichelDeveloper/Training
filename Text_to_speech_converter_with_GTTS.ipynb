```python
#import necessary library
from gtts import gTTS
from IPython.display import Audio
```

```python
#create speech file
text_to_say = "I'm saying what was typed in the text to say variable"
language = "en"
gtts_object = gTTS(text = text_to_say,
     lang = language,
     slow = False)

#save file
gtts_object.save("/content/gtts.wav")
```

```python
#read file
Audio("/content/gtts.wav")
```

```python
#create speech file
french_text = "ceci est un texte dit en fran√ßais"
french_language = "fr"
gtts_object = gTTS(text = french_text,
     lang = french_language,
     slow = True)

#save file
gtts_object.save("/content/gtts.wav")

#read file
Audio("/content/gtts.wav")
```
